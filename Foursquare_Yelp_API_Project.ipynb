```python
def get_tables_yelp(M):
    COL_N=[]
    COL_R=[]
    COL_RC=[]
    for i in M['businesses']:
        COL_N.append(i['name'])
        COL_R.append((i['rating'])*2)
        COL_RC.append(i['review_count'])

        
    YEP={'name':COL_N , 'rating (2X scale)': COL_R, 'review_count':COL_RC}
    TABLE=pd.DataFrame(data=YEP).sort_values(by= 'rating (2X scale)', ascending=False)
    return(TABLE)
```

```python
def Wonho_recommend_score(input_df):
    try:
        input_df['popularity_multiplier']=1+(input_df['count'])/(input_df['count'].max())
    except:
        input_df['popularity_multiplier']=1+(input_df['review_count'])/(input_df['review_count'].max())
    
    try:
        input_df['Wonho_recommend_score']=(input_df['popularity_multiplier'])*(input_df['rating'])
    except:
        input_df['Wonho_recommend_score']=(input_df['popularity_multiplier'])*(input_df['rating (2X scale)'])
    return input_df
```
